# bar chart

```python
 
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
plt.figure(figsize=(16,6))
plt.bar(classes,class1_students,width=0.3,align="edge",color="y",
       edgecolor="m" ,linewidth= 1,alpha = 0.7,
           linestyle= "--",label="class 1 student",visible=True)
plt.legend()
```

```python
plt.figure(figsize=(16,6))
plt.barh(classes,class1_students,align="edge",color="y",
       edgecolor="m" ,linewidth= 1,alpha = 0.7,
           linestyle= "--",label="class 1 student",visible=True)
plt.legend()
```

```python
plt.figure(figsize=(16,6))
plt.bar(classes,class1_students,width=0.3,align="edge",color="y",
       edgecolor="m" ,linewidth= 1,alpha = 0.7,
           linestyle= "--",label="class 1 student",visible=True)
plt.title(" Bar chart of class",fontsize=18)
plt.xlabel("Classes",fontsize=12)
plt.ylabel("No of Students",fontsize=12)
plt.legend()
```

```python
plt.figure(figsize=(16,6))
plt.bar(classes,class1_students,width=0.3,align="edge",color="y",
           linestyle= "--",label="class 1 student",visible=True)
plt.bar(classes,class2_students,width=0.3,align="edge",color="g",
           linestyle= "-",label="class 2 student",visible=True)
plt.title(" Bar chart of class",fontsize=18)
plt.xlabel("Classes",fontsize=12)
plt.ylabel("No of Students",fontsize=12)
plt.legend()
```

```python
plt.figure(figsize=(16,6))
classes_index=np.arange(len(classes))
width=0.3
plt.bar(classes_index,class1_students,width,align="edge",color="y",
           linestyle= "--",label="class 1 student",visible=True)
plt.bar(classes_index+width,class2_students,width,align="edge",color="g",
           linestyle= "-",label="class 2 student",visible=True)
plt.bar(classes_index+width*2,class3_students,width,align="edge",color="r",
           linestyle= "-",label="class 2 student",visible=True)
plt.xticks(classes_index+width,classes,rotation=20)
plt.title(" Bar chart of class",fontsize=18)
plt.xlabel("Classes",fontsize=12)
plt.ylabel("No of Students",fontsize=12)
plt.legend()
```

```python
plt.figure(figsize=(16,6))
classes_index=np.arange(len(classes))
width=0.3
plt.barh(classes_index,class1_students,width,align="edge",color="y",
           linestyle= "--",label="class 1 student",visible=True)
plt.barh(classes_index+width,class2_students,width,align="edge",color="g",
           linestyle= "-",label="class 2 student",visible=True)
plt.barh(classes_index+width*2,class3_students,width,align="edge",color="r",
           linestyle= "-",label="class 2 student",visible=True)
plt.yticks(classes_index+width,classes,rotation=20)
plt.title(" Bar chart of class",fontsize=18)
plt.ylabel("Classes",fontsize=12)
plt.xlabel("No of Students",fontsize=12)
plt.legend()
```

# pie chart

```python

classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_students,labels = classes)
plt.show()
```

```python
explode = [0.08,0,0.1,0,0]
colors = ["c","b","r","y","g"]
wedgeprops = {"linewidth" : 4 ,"width":2, "edgecolor":"b"}
plt.pie(class1_students,labels = classes,explode=explode,
        colors= colors , autopct = "%0.1f%%",shadow=True,radius=1.9,
       startangle=90,textprops={"fontsize":15},pctdistance=0.5,
        labeldistance = 1.1,counterclock= False,wedgeprops=wedgeprops,frame=False,rotatelabels=3)#center=(2,3)
plt.show()
```

```python
explode = [0.08,0,0.1,0,0]
colors = ["c","b","r","y","g"]

plt.pie(class1_students,labels = classes,explode=explode,
        colors= colors , autopct = "%0.1f%%",shadow=True,radius=1.9,
       startangle=90,textprops={"fontsize":15},pctdistance=0.5)
plt.legend()
plt.show()
```
