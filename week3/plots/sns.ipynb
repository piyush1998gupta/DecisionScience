```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df = pd.read_csv("Admission_Predict.csv",index_col="Serial No.")
df = df.iloc[:,:]
```

```python
plt.figure(figsize=(20,7))
```

## histogram

```python
admitted = df[df['Research']==1]
not_admitted = df[df['Research']==0]
sns.distplot(a=admitted['University Rating'],label="Admitted",kde=False)
sns.distplot(a=not_admitted['University Rating'],label="not admitted",kde=False)
plt.title("Admitted Vs Not Admitted Students in University")
plt.legend()
```

```python
sns.kdeplot(data=admitted['University Rating'],label = "Admitted",shade=True)
sns.kdeplot(data=not_admitted['University Rating'],label="not Admitted",shade=True)
```

```python
# background changes
sns.set_style("darkgrid") # (1)"darkgrid", (2)"whitegrid", (3)"dark", (4)"white", and (5)"ticks"
sns.kdeplot(data=admitted['University Rating'],label = "Admitted",shade=True)
```
